## The welcome page

### the quick start section



It contains some shortcuts to start JupyterLab or to create a new plot panel.

 ***

### the workspace section



This section provides a centralized view of all your SciQLop workspaces. Think of a workspace as your personal science sandbox! Each workspace is a dedicated directory where you can:

* **Craft Jupyter Notebooks**: Build interactive analysis notebooks to explore and visualize your space physics data.
* **Write Python Scripts**: Develop custom Python scripts to tailor SciQLop's functionalities to your specific needs.
* **Organize Your Data**: Store and manage your space physics data within your workspace for easy access.
* **Manage Dependencies**: Install and manage any additional Python packages required for your unique analyses.

 ***

### the examples section



The example section provides a library of pre-built templates to jumpstart your space physics analysis.  These examples can include:

* **Jupyter Notebooks**: Ready-made notebooks showcasing common analysis workflows for various space physics datasets.
* **Python Scripts**: Custom scripts demonstrating advanced functionalities or tailored data manipulations.
* **Data Files**: Datasets relevant to the example, conveniently stored within the workspace.
* **Dependencies**: A list of any additional Python packages required to run the example, automatically installed for your convenience.

Loading an example offers two options:

1. **New Workspace**: If you don't have a currently loaded workspace, the example will create a new one populated with its contents.
2. **Merge into Workspace**: If you already have a workspace loaded, the example will merge its contents (notebooks, scripts, data, and dependencies) into your existing workspace.

 ***

## The time range widget



The time range widget in the top toolbar is your control center for navigating your space physics data. It allows you to:

* **Define Default Interval**: Set the starting and ending points for your initial data view. This default time range will be used in various functionalities throughout SciQLop, such as when creating new plots.
* **Drag-and-Drop Power**: Drag and drop the time range widget directly onto a plot panel to instantly set the x-axis range for that specific plot.
 
 ***

## The porducts panel



The product panel acts as the gateway to your space physics data exploration journey. It's organized like a hierarchical tree structure, where leaves are plottable products. The main sources for this tree are:

* **Extensive Speasy Integration**: SciQLop seamlessly integrates with Speasy, a powerful tool that grants access to a vast collection of real space physics measurements from various spacecraft missions and instruments.
* **User-Defined Virtual Products**: Beyond pre-loaded data, SciQLop empowers you to define your own custom "virtual products." These are essentially functions that can combine or manipulate existing data to create new physical quantities tailored to your specific research questions.

 ***

## The catalog quick access



Event Catalog at Your Side

The catalog quick access panel is your plot's companion for efficiently managing and exploring event catalogs. It provides a simplified view compared to the full catalog interface, offering key functionalities directly within your plotting workflow:

* **Event Visualization**: Effortlessly overlay events from your chosen catalog onto your current plot for a combined view of data and identified events.
* **Seamless Navigation**: Jump from event to event within the catalog using intuitive controls, allowing you to quickly investigate specific events of interest.
* **Event Management Power**: This quick access panel empowers you to create, edit, and even delete events directly within your SciQLop workspace, streamlining your event catalog management.

 ***

## The inspector widget



The inspector widget acts as your personal investigator for all your plots and visualizations. It provides a comprehensive tree view, allowing you to:

* **Explore Plot Hierarchy**: Navigate through a detailed breakdown of your SciQLop workspace, including plot panels, individual plots, and underlying graphs.
* **Effortless Customization**: The inspector empowers you to fine-tune the appearance and behavior of your visualizations directly. Simply select an element in the tree view and adjust its properties, such as switching the y-axis scale to logarithmic.
* **Streamlined Management**: Need to declutter your workspace? The inspector allows you to delete plots, panels, or graphs with ease, keeping your visualizations organized and focused.

 ***

SciQLop empowers you to explore your data visually with intuitive mouse wheel interactions:

* **Panning**: Simply scroll the mouse wheel **within the plot area** to pan horizontally across your data. This allows you to smoothly navigate through different time intervals.
* **Zooming** on an Axis: Hover your mouse cursor over a specific axis (X or Y) and scroll the wheel. Zooming **in** will magnify the data displayed on that axis, allowing you to focus on specific details. Scrolling the wheel in the opposite direction will zoom **out**, providing a broader view of your data.
* **Advanced Zooming**: Hold down specific keys while scrolling the wheel to achieve additional zooming functionalities:
  * **CTRL + Wheel**: Zoom **horizontally** within the plot area. This offers a shortcut for panning large datasets.
  * **SHIFT + Wheel**: Zoom **vertically** on the specific axis under your cursor. This allows you to focus on features along a particular data component.
