# SciQLop GUI tour



## The welcome page

The welcome page is your starting point when you open SciQLop. It features a global filter bar at the top and is organized into several sections.

### The quick start section



The quick start row contains shortcut cards to common actions such as creating a new plot panel or opening JupyterLab.

 ***

### The recent workspaces section



This section shows your recently used SciQLop workspaces as a card grid. Think of a workspace as your personal science sandbox! Each workspace is a dedicated directory where you can:

* **Craft Jupyter Notebooks**: Build interactive analysis notebooks to explore and visualize your space physics data.
* **Write Python Scripts**: Develop custom Python scripts to tailor SciQLop's functionalities to your specific needs.
* **Organize Your Data**: Store and manage your space physics data within your workspace for easy access.
* **Manage Dependencies**: Install and manage any additional Python packages required for your unique analyses.

 ***

### The examples section



The examples section provides a library of pre-built examples to jumpstart your space physics analysis. Each example is a self-contained bundle that can include Jupyter notebooks, Python scripts, data files, and a list of required dependencies.

When you open an example, its contents are copied into a subfolder of your current workspace. If the example requires additional Python packages, you will be prompted to install them automatically.

 ***

## The per-panel time range bar



Each plot panel has its own time range bar at the bottom. It allows you to:

* **Navigate in time**: Adjust the start and end times for the data displayed in that specific panel.
* **Step through time**: Use the navigation buttons to step forward or backward in time.
* **Set duration**: Control the time span displayed in the panel.

 ***

## The products panel



The product panel acts as the gateway to your space physics data exploration journey. It is organized like a hierarchical tree structure, where leaves are plottable products. The main sources for this tree are:

* **Extensive Speasy Integration**: SciQLop seamlessly integrates with Speasy, a powerful tool that grants access to a vast collection of real space physics measurements from various spacecraft missions and instruments.
* **User-Defined Virtual Products**: Beyond pre-loaded data, SciQLop empowers you to define your own custom "virtual products." These are essentially functions that can combine or manipulate existing data to create new physical quantities tailored to your specific research questions.

You can drag and drop products from the tree directly onto a plot panel, or use the search overlay that appears when creating a new panel.

 ***

## The catalog browser



The catalog browser is your companion for managing and exploring event catalogs. It features a tree view of catalog providers on top and an event table below, offering:

* **Event Visualization**: Overlay events from your chosen catalog onto your current plot for a combined view of data and identified events.
* **Seamless Navigation**: Jump from event to event within the catalog, allowing you to quickly investigate specific events of interest.
* **Event Management**: Create, edit, and delete events directly within the catalog browser.
* **Filtering**: Filter catalogs by name to quickly find what you need.
* **Save indicators**: Dirty (unsaved) providers are visually flagged so you know when to save.

 ***

## The properties panel



The properties panel lets you examine and customize your plots and visualizations. It provides a tree view allowing you to:

* **Explore Plot Hierarchy**: Navigate through a detailed breakdown of your plot panels, individual plots, and underlying graphs.
* **Effortless Customization**: Fine-tune the appearance and behavior of your visualizations directly. Simply select an element in the tree view and adjust its properties, such as switching the y-axis scale to logarithmic.
* **Streamlined Management**: Delete plots, panels, or graphs to keep your visualizations organized and focused.

 ***

SciQLop empowers you to explore your data visually with intuitive mouse wheel interactions:

* **Panning**: Simply scroll the mouse wheel **within the plot area** to pan horizontally across your data. This allows you to smoothly navigate through different time intervals.
* **Zooming** on an Axis: Hover your mouse cursor over a specific axis (X or Y) and scroll the wheel. Zooming **in** will magnify the data displayed on that axis, allowing you to focus on specific details. Scrolling the wheel in the opposite direction will zoom **out**, providing a broader view of your data.
* **Advanced Zooming**: Hold down specific keys while scrolling the wheel to achieve additional zooming functionalities:
  * **CTRL + Wheel**: Zoom **horizontally** within the plot area. This offers a shortcut for panning large datasets.
  * **SHIFT + Wheel**: Zoom **vertically** on the specific axis under your cursor. This allows you to focus on features along a particular data component.
