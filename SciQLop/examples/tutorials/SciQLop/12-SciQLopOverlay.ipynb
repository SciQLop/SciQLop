Every plot in SciQLop has an **overlay** — an in-canvas message banner you can use to surface status information, warnings, or errors on top of the data, without adding a separate widget.

**What you'll learn**
- Show, clear, and inspect the overlay message of a plot.
- Choose a severity level (Info / Warning / Error) to drive the visual style.
- Position and size the overlay; make it collapsible; adjust its opacity.

**Prerequisites** — [Plot panels](5-SciQLopBasicPlotPanel.ipynb).

**Next** — [Graphic primitives](10-SciQLopGraphicPrimitives.ipynb), [Annotation layers](16-SciQLopAnnotationLayers.ipynb).


## 1. Showing a message

Access the overlay via `plot.overlay` and call `.show(text)` to display a message.

```python
from SciQLop.user_api.plot import create_plot_panel, OverlayLevel, OverlaySizeMode, OverlayPosition
from SciQLop.user_api import TimeRange

p = create_plot_panel()
p.time_range = TimeRange("2015-11-18T02:14:30", "2015-11-18T03:34:00")

plot, _ = p.plot("speasy/cda/MMS/MMS1/FGM/MMS1_FGM_SRVY_L2/mms1_fgm_b_gse_srvy_l2")

plot.overlay.show("Survey data — 16 vectors/sec")
```

## 2. Severity levels

Messages have a severity level that controls their visual style.

```python
plot.overlay.show("Data gap detected", level=OverlayLevel.Warning)
```

```python
plot.overlay.show("Fetch failed: timeout", level=OverlayLevel.Error)
```

```python
plot.overlay.clear()
```

## 3. Position and size mode

Control where the overlay appears and how it sizes itself.

```python
plot.overlay.show(
    "Bottom-anchored, compact overlay",
    position=OverlayPosition.Bottom,
    size_mode=OverlaySizeMode.Compact
)
```

```python
plot.overlay.show(
    "Full-width overlay at the top",
    position=OverlayPosition.Top,
    size_mode=OverlaySizeMode.FullWidget
)
```

```python
plot.overlay.show(
    "FitContent grows the banner to fit its text and no more",
    position=OverlayPosition.Top,
    size_mode=OverlaySizeMode.FitContent,
)
```

## 4. Collapsible overlays

Make overlays user-dismissable without clearing them entirely.

```python
plot.overlay.show("Click to collapse this message")
plot.overlay.collapsible = True

# Collapse/expand programmatically
plot.overlay.collapsed = True
plot.overlay.collapsed = False
```

## 5. Opacity

Adjust transparency so the overlay doesn't obscure data.

```python
plot.overlay.show("Semi-transparent")
plot.overlay.opacity = 0.5
```

## 6. Reading overlay state

```python
print(f"Text:      {plot.overlay.text}")
print(f"Level:     {plot.overlay.level}")
print(f"Position:  {plot.overlay.position}")
print(f"Size mode: {plot.overlay.size_mode}")
print(f"Opacity:   {plot.overlay.opacity}")
```

```python
plot.overlay.clear()
```

## API reference

| Property / Method | Description |
|---|---|
| `plot.overlay.show(text, *, level, size_mode, position)` | Display a message |
| `plot.overlay.clear()` | Remove the message |
| `.text`, `.level`, `.position`, `.size_mode` | Read current state |
| `.collapsible` | Toggle user-collapsible behaviour |
| `.collapsed` | Programmatically collapse/expand |
| `.opacity` | Transparency (0.0 – 1.0) |

| Enum | Values |
|---|---|
| `OverlayLevel` | `Info`, `Warning`, `Error` |
| `OverlaySizeMode` | `Compact`, `FitContent`, `FullWidget` |
| `OverlayPosition` | `Top`, `Bottom`, `Left`, `Right` |
